```
import tensorflow as tf
tf.set_random_seed(777)

# traning set.1   x*w + random value(int 0-2)
x_train = [2,1,6,2,6,10,7,2]
y_train = [5,2,12,6,13,21,16,5]

# traning set.2  x*w + x*w
x_train = [8,10,3,3,5,0,5,1]
y_train = [21,26,8.5,8.5,13.5,1,13.5,3.5]


# Try to find values for W and b to compute Y = W * X + b
W = tf.Variable(tf.random_normal([1]), name="weight")
b = tf.Variable(tf.random_normal([1]), name="bias")

# placeholders for a tensor that will be always fed using feed_dict
# See http://stackoverflow.com/questions/36693740/
X = tf.placeholder(tf.float32, shape=[None])
Y = tf.placeholder(tf.float32, shape=[None])

# Our hypothesis is X * W + b (가설)
hypothesis = X * W + b

# cost/loss function (비용)
cost = tf.reduce_mean(tf.square(hypothesis - Y))

# optimizer (비용을 줄이는 방향으로 계산및 이동처리)
train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)

# Launch the graph in a session.
with tf.Session() as sess:
    # Initializes global variables in the graph.
    sess.run(tf.global_variables_initializer())

    # Fit the line
    for step in range(2001):
        _, cost_val, W_val, b_val = sess.run([train, cost, W, b], feed_dict={X:x_train , Y:y_train})
        if step % 20 == 0:
            print(step, cost_val, W_val, b_val)

    # Testing our model
    print(sess.run(hypothesis, feed_dict={X: [5]}))
    print(sess.run(hypothesis, feed_dict={X: [2.5]}))
    print(sess.run(hypothesis, feed_dict={X: [1.5, 3.5]}))
```
